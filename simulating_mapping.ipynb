```python
# --- Simulating classified EEG output ---

#2. Simulating EEG Output
# def simulate_eeg_command():
#     """
#     Simulating EEG command output
#     """
#     # Option 1: Left/Right Only 
#     commands = ['move_left', 'move_right']
    
#     # Option 2: Left/Right/Up/Down 
#     # commands = ['move_left', 'move_right', 'move_up', 'move_down'] #uncomment to use (& comment other commands)
    
#     # Option 3: Left/Right/Click/Double Click 
#     # commands = ['move_left', 'move_right', 'click', 'double_click'] #uncomment to use (& comment other commands)
    
#     return random.choice(commands)






# --- Simulating recording & classification ---

def simulate_eeg_recording(duration=1.0, sampling_rate=256):
    """
    Simulates EEG recording for a given duration & sampling rate
    Returns a list of synthetic EEG values
    
    As demonstration: we generate a sine wave at 10 Hz with added Gaussian noise
    """
    n_samples = int(duration * sampling_rate)
    eeg_data = [math.sin(2 * math.pi * 10 * t / sampling_rate) + random.gauss(0, 0.5) 
                for t in range(n_samples)]
    return eeg_data

def process_and_classify_eeg(eeg_data):
    """
    Processes EEG data & classify it into a command
    
    Note: placeholder for an actual EEG processing pipeline
    As demonstration:
      - We calculate average of the EEG data
      - If average is above 0, we classify the command as 'move_right'
      - Otherwise, we classify it as 'move_left'
    """
    avg_value = sum(eeg_data) / len(eeg_data)
    threshold = 0.0  #can adjust theshold based on our actual data characteristics
    if avg_value > threshold:
        return 'move_right'
    else:
        return 'move_left'
    
def simulate_eeg_command():
    """
    Simulates EEG command output by:
      1. Recording a segment of synthetic EEG data
      2. Processing & classifying that data to determine a command
    
    Returns:
        A command string (e.g., 'move_left' or 'move_right')
    """
    eeg_data = simulate_eeg_recording(duration=1.0, sampling_rate=256)
    command = process_and_classify_eeg(eeg_data)
    return command





# --- Mapping EEG Commands to Cursor Actions ---

def execute_command(command):
    """
    Mapping command (from simulated or real EEG) to an actual cursor action
    """
    #getting current position of cursor
    current_x, current_y = pyautogui.position()
    
    #Option 1: Left/Right Only 
    if command == 'move_left':
        print("Moving left")
        new_x = max(current_x - 200, 0)
        pyautogui.moveTo(new_x, current_y, duration=1)  #movement over 1 second
    elif command == 'move_right':
        print("Moving right")
        screen_width, _ = pyautogui.size()
        new_x = min(current_x + 200, screen_width)
        pyautogui.moveTo(new_x, current_y, duration=1)
    else:
        print("Unknown command:", command)
    
    #Option 2: Left/Right/Up/Down 
    #uncomment following to use
    """
    if command == 'move_left':
        print("Moving left")
        new_x = max(current_x - 200, 0)
        pyautogui.moveTo(new_x, current_y, duration=1)
    elif command == 'move_right':
        print("Moving right")
        screen_width, _ = pyautogui.size()
        new_x = min(current_x + 200, screen_width)
        pyautogui.moveTo(new_x, current_y, duration=1)
    elif command == 'move_up':
        print("Moving up")
        new_y = max(current_y - 200, 0)
        pyautogui.moveTo(current_x, new_y, duration=1)
    elif command == 'move_down':
        print("Moving down")
        screen_height = pyautogui.size().height
        new_y = min(current_y + 200, screen_height)
        pyautogui.moveTo(current_x, new_y, duration=1)
    else:
        print("Unknown command:", command)
    """
    
    #Option 3: Left/Right/Click/Double Click 
    #uncomment following to use
    """
    if command == 'move_left':
        print("Moving left")
        new_x = max(current_x - 200, 0)
        pyautogui.moveTo(new_x, current_y, duration=1)
    elif command == 'move_right':
        print("Moving right")
        screen_width, _ = pyautogui.size()
        new_x = min(current_x + 200, screen_width)
        pyautogui.moveTo(new_x, current_y, duration=1)
    elif command == 'click':
        print("Clicking")
        pyautogui.click()
    elif command == 'double_click':
        print("Double clicking")
        pyautogui.doubleClick()
    else:
        print("Unknown command:", command)
    """
```
